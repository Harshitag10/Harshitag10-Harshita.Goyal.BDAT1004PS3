```python
#step 1
import numpy as np
import pandas as pd
```

```python
#step 2 and 3
#importing dataset to variable data and replacing the first 3 columns by a proper datetime index
data=pd.read_csv("wind.txt",sep="\s+")
data.head()
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
#step 4 and 5
#creating a function to fix the year and setting dates aas index
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,data.Date,data.Date - pd.offsets.DateOffset(years=100))

newData = data.set_index("Date")
newData.index.astype("datetime64[ns]")
```

```python
#step 6 - finding missing values

print(newData.isnull().sum().sum())
```

```python
#step 7 - finding total number of non missing values

print(newData.count().sum())
```

```python
#step 8- calculating mean windspeeds

newData.mean().mean()
```

```python
#step 9- Creating a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations at each location over all the days


def calc(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(calc)
loc_stats
```

```python
#step 10 - Creating a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day

day_stats = newData.apply(calc,axis=1)
day_stats.head(10)
```

```python
#step 11 - finding average windspeed in January for each location

jandata = newData[newData.index.month == 1]
print (jandata.mean().round(2))
```

```python
#step 12- yearly frequency of each location

print(newData.resample('A').mean())
```

```python
#step 13- monthly frequency of each location

print(newData.resample('M').mean())
```

```python
#step 14- weekly frequency of each location

print(newData.resample('W').mean())
```

```python
# step 15 - finding stats over all locations over the year

yeardata = newData[newData.index.year == 1961]
yearstats = newData.resample('W').mean().apply(lambda x: x.describe())
print (yearstats)
```
